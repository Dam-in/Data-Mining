```python
import pandas as pd

data = pd.read_csv('/Users/damin/Desktop/iris.csv')
data
```

```python
# 1단계
import itertools
import matplotlib.pyplot as plt
```

```python
# 2단계
features = [ 'SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth' ]
combinations = itertools.combinations(features, 2)
```

```python
setosaData = data[data['Name'] == 'Iris-setosa']
virginicaData = data[data['Name'] == 'Iris-virginica']
versicolorData = data[data['Name'] == 'Iris-versicolor']
group = setosaData, virginicaData, versicolorData
```

```python
figure_index = 1
combinations = itertools.combinations(features, 2)
fig = plt.figure(figsize=(15, 10))
for combination in combinations:
    axs = fig.add_subplot(2,3,figure_index)


    for group_df in group:
        axs.scatter(group_df[combination[0]], group_df[combination[1]], label = group_df['Name'].iloc[0])

    axs.set_xlabel(combination[0])
    axs.set_ylabel(combination[1])

    figure_index+=1

plt.legend()
plt.show()
```

```python
# 3단계
plt.figure(figsize=(10,10))


for group_df in group:
    plt.scatter(group_df['PetalLength'], group_df['PetalWidth'], label = group_df['Name'].iloc[0])

plt.fill([0.8, 0.8, 2.2, 2.2], [0.0, 0.7, 0.7, 0.0], color='Blue', alpha=0.1)
plt.fill([2.8, 2.8, 4.8, 4.8], [0.9, 1.7, 1.7, 0.9], color='Green', alpha=0.1)
plt.fill([4.8, 4.8, 7.0, 7.0], [1.5, 2.6, 2.6, 1.5], color='Orange', alpha=0.1)
    

plt.minorticks_on()
plt.grid(which='both')

plt.xlabel("Petal length")
plt.ylabel("Petal width")

plt.legend()
plt.show()
```
