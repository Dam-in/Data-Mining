```python
# 상관관계 조사
corr_matrix = data.corr()
corr_matrix['Revenue'].sort_values(ascending=False)
# 1에 가까우면 강한 양의 상관관계를 가진다는 뜻
# -1에 가까우면 강한 음의 상관관계를 가진다는 뜻
# 0에 가까우면 선형적인 상관관계가 없다는 뜻
```
